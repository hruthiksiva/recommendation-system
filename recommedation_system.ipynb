```python
all_genres = set()
for genres in df['genre']:
  if isinstance(genres, str):
    for g in genres.split(','):
      all_genres.add(g)
```

```python
all_genres
```

```python
for genre in all_genres:
  df[genre] = df['genre'].apply(lambda x:1 if isinstance(x, str) and genre in x.split(',') else 0)
df
```

```python
from sklearn.metrics.pairwise import cosine_similarity

feature_vectors = df[list(all_genres)]
```

```python
cosine_sim = cosine_similarity(feature_vectors)
```

```python
cosine_sim[0][1]
```
